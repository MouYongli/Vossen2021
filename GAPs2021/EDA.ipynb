# Exploratory Data Analysis on GAP dataset

```python
import os
import os.path as osp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

## Mask images are stored as grey value images.
## Coding of the grey values:
##    0 = VOID
##    1 = intact road,
##    2 = applied patch,
##    3 = pothole,
##    4 = inlaid patch,
##    5 = open joint,
##    6 = crack
##    7 = street inventory

here = os.getcwd()
destination_directory = here
basedir = os.path.join(destination_directory, 'v2', 'segmentation')
images_dir = os.path.join(basedir, 'images')
train_dir = os.path.join(basedir, 'train')
valid_dir = os.path.join(basedir, 'valid')
test_dir = os.path.join(basedir, 'test')
validtest_dir = os.path.join(basedir, 'valid-test')
list_images = os.listdir(images_dir)
list_train = os.listdir(train_dir)
list_valid = os.listdir(valid_dir)
list_test = os.listdir(test_dir)
list_validtest = os.listdir(validtest_dir)
images_df = pd.DataFrame(np.array([list_images, list_images]).T, columns=['id','img_name'])
train_df = pd.DataFrame(np.array([list_train, list_train]).T, columns=['id','img_name'])
valid_df = pd.DataFrame(np.array([list_valid, list_valid]).T, columns=['id','img_name'])
test_df = pd.DataFrame(np.array([list_test, list_test]).T, columns=['id','img_name'])
validtest_df = pd.DataFrame(np.array([list_validtest, list_validtest]).T, columns=['id','img_name'])
images_df = images_df[images_df['id'].str.endswith('.jpg')]
train_df = train_df[train_df['id'].str.endswith('.png')]
valid_df = valid_df[valid_df['id'].str.endswith('.png')]
test_df = test_df[test_df['id'].str.endswith('.png')]
validtest_df = validtest_df[validtest_df['id'].str.endswith('.png')]
```

```python
images_df['id'] = images_df['img_name'].map(lambda x: x[:-4])
train_df['id'] = train_df['img_name'].map(lambda x: x[:-4])
valid_df['id'] = train_df['img_name'].map(lambda x: x[:-4])
test_df['id'] = train_df['img_name'].map(lambda x: x[:-4])
validtest_df['id'] = validtest_df['img_name'].map(lambda x: x[:-4])
```

## Class Distribution on Training Dataset

```python
import cv2
# mask = cv2.imread(osp.join(train_dir, train_df['img_name'][0]), cv2.IMREAD_GRAYSCALE)
# plt.imshow(mask)
# plt.show()
class_names = ['void', 'intact_road', 'applied_patch', 'pothole', 'inlaid_patch', 'open_joint', 'crack', 'street_inventory']
cls_dist = pd.DataFrame(columns=['image', 'void', 'intact_road', 'applied_patch', 'pothole', 'inlaid_patch', 'open_joint', 'crack', 'street_inventory'])
count = 0
for img_name in train_df['img_name']:
    print(img_name)
    mask = cv2.imread(osp.join(train_dir, img_name), cv2.IMREAD_GRAYSCALE)
    cls_dist.loc[count] = [img_name] + [(mask == i).sum() for i in range(len(class_names))]
    count += 1
cls_dist.to_csv('train_cls_dist.csv')
for cls in class_names:
    plt.figure()
    plt.title(cls)
    cls_dist[cls].plot.hist()
print(cls_dist.mean(axis=0))
cls_dist.mean(axis=0).plot.bar()
```
